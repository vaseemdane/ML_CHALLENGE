```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
import numpy as np
```

```python
data=fetch_california_housing(as_frame=True)
df=data.frame
```

```python
x=df.drop(['MedHouseVal'],axis=1)
y=df['MedHouseVal']
```

```python
def cost(x,y,w,b):
    total_cost=0
    m=x.shape[0]
    for i in range(m):
        f_wb=np.dot(x.iloc[i],w)+b
        total_cost+=(f_wb-y[i])**2
    total_cost=total_cost/(2*m)
    return total_cost
```

```python
w=np.zeros(x.shape[1])
b=0
cost(x,y,w,b)
```

```python
def gradient_function(x,y,w,b):
    m,n=x.shape
    dj_dw=np.zeros((n,))
    dj_db=0
    for i in range(m):
        error=(np.dot(x.iloc[i],w)+b)-y[i]
        for j in range(n):
            dj_dw[j]=dj_dw[j]+error*x.iloc[i,j]
        dj_db=dj_db+error
    dj_dw=dj_dw/m
    dj_db=dj_db/m
    return dj_dw,dj_db        
```

```python
gradient_function(x,y,w,b)
```

```python
def gradient_descent(x,y,w,b,cost,gradient_function,alpha,iter):
    j_history=[]
    w_history=[]
    m,n=x.shape
    for i in range(iter):
        dj_dw,dj_db=gradient_function(x,y,w,b)
        w=w-(alpha/m)*dj_dw
        b=b-(alpha/m)*dj_db

        cost_values=cost(x,y,w,b)
        j_history.append(cost_values)
        w_history.append(w)
        if i%10==0:
            print(f'the cost at iteration {i} is {j_history[-1]:.2f}')
    return w,b,j_history,w_history        
```

```python
alpha=0.001
iter=100
w,b,j_history,_=gradient_descent(x,y,w,b,cost,gradient_function,alpha,iter)
print(f'the final values of w and b are {w} and {b}')
```
